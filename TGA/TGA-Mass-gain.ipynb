```python
import numpy as np
import pandas as pd
df = pd.read_csv('PDC_LDFZ_20200720_1_217_(PTF1_20200715)-SMB_TGA_underO2.txt',delimiter = '\t',header = 48, names = ['Time','Temp','Weight','Heat Flow','Temp diff','Temp diff V','Purge flow'])
df
```

```python
%matplotlib
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Time (min)')
ax1.set_ylabel('Weight (mg)', color=color)
ax1.plot(df['Time'][:-1], df['Weight'][:-1], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Temperature ($^{\circ}$C)', color=color)  # we already handled the x-label with ax1
ax2.plot(df['Time'][:-1], df['Temp'][:-1], color=color, linestyle = '--')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
plt.savefig('TGA-BWO.png',dpi=200)
```
